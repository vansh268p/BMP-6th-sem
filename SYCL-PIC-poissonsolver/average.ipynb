```python
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import re

# === Directory containing Ex data ===
base_dir = "out/Ex/"
file_pattern = os.path.join(base_dir, "*")
files = sorted(glob.glob(file_pattern), key=lambda f: int(re.search(r'\d+', os.path.basename(f)).group()))

# === Simulation parameters ===
grid_size = 257    # Assuming a 200x200 grid
mid_row = int(grid_size / 2)  # 100th row for y = 0.5 m if the domain is 1m x 1m
dx = 1.0 / grid_size
time_step = 2.5e-9
time_scale = 1e6  # Convert to microseconds

# === Extract and plot Ex over time ===
time_points = []
Ex_profiles = []

for file in files:
    # Extract time from the filename
    iteration = int(re.search(r'\d+', os.path.basename(file)).group())
    time_us = iteration * time_step * time_scale
    time_points.append(time_us)
    
    # Load the Ex data and extract the middle row
    data = np.loadtxt(file)
    Ex_profiles.append(data[mid_row, :])

# Convert to numpy arrays for easier plotting
time_points = np.array(time_points)
Ex_profiles = np.array(Ex_profiles)
Ex_profiles = np.flipud(Ex_profiles[0:250])

# === Plot ===
plt.figure(figsize=(6, 6))
plt.imshow(Ex_profiles, aspect='auto', extent=[time_points[0], time_points[-1], 0, 2], cmap='seismic', vmin=-10, vmax=10)
#plt.title(r"$E_x$ along $y = 0.5$ m over time")
plt.xlabel(r"$x$ [m]")
plt.ylabel(r"$t$ [µs]")
cbar = plt.colorbar(label=r"$E_x$ [V/m]")
cbar.set_ticks(np.linspace(-10, 10, 9))
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import glob
import re

plt.rcParams.update({'font.size': 16})

# === Natural sort helper ===
def natural_key(s):
    return [int(text) if text.isdigit() else text.lower() for text in re.split(r'(\d+)', os.path.basename(s))]

# === Electron and Ion Densities ===
def load_density(folder):
    files = sorted(glob.glob(os.path.join(folder, "*")), key=natural_key)
    time, avg = [], []
    for f in files:
        try:
            t = int(re.search(r'(\d+)', os.path.basename(f)).group(1)) * 2.5e-9 * 1e6
            d = np.loadtxt(f)
            time.append(t)
            avg.append(np.mean(d))
        except:
            continue
    return np.array(time), np.array(avg) / avg[0]

# === Load benchmark densities ===
bench_dens = pd.read_csv("DensityTemporalProfiles_LPP.csv", sep=",", decimal=",")
x_bench = bench_dens['t'].values.astype(float)
y_ne = bench_dens['case2_ne_vki'].values.astype(float)
y_ni = bench_dens['case2_ni_vki'].values.astype(float)

# === Load all densities ===
t_ne, ne_avg = load_density("out_0/200x200_2500000_2001/electronDensity/")
t_ni, ni_avg = load_density("out_0/200x200_2500000_2001/ionDensity/")

# === Load energies ===
def load_energy(file, scale=1.0):
    data = np.loadtxt(file)
    return data[:, 0], data[:, 1] * scale

scale_ev_to_j = 1.602e-19

it_ion_kin, ion_kin = load_energy("out_0/200x200_2500000_2001/TotalIonKinEnergy.out", scale_ev_to_j)
it_ion_pot, ion_pot = load_energy("out_0/200x200_2500000_2001/TotalIonPotEnergy.out", scale=1/2.0)
it_elec_kin, elec_kin = load_energy("out_0/200x200_2500000_2001/TotalElecKinEnergy.out", scale_ev_to_j)
it_elec_pot, elec_pot = load_energy("out_0/200x200_2500000_2001/TotalElecPotEnergy.out", scale=1/2.0)

time_us = it_ion_kin * 2.5e-9 * 1e6
total_energy = ion_kin + ion_pot + elec_kin + elec_pot

# === Load benchmark energy ===
bench = np.loadtxt("ResultsComparison/IntegratedQuantities/data_case2_lpp/data_case_01.dat")
bench_time = bench[:, 0] * 1e6
bench_elec_kin = bench[:, 2] * scale_ev_to_j
bench_elec_pot = bench[:, 3] * -scale_ev_to_j
bench_ion_kin = bench[:, 5] * scale_ev_to_j
bench_ion_pot = bench[:, 6] * scale_ev_to_j
bench_total_energy = bench_elec_kin + bench_elec_pot + bench_ion_kin + bench_ion_pot

# === Plot all 4 graphs ===
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
titles = ['(a) ', '(b) ', '(c) ', '(d)']
fontsize = 23

# (1) Electron + Ion Density
axs[0, 0].plot(t_ne, ne_avg, 'r-', label="Ele. In house PIC-MCC")
axs[0, 0].plot(x_bench, y_ne, 'r--', label="Electron LPP")
axs[0, 0].plot(t_ni, ni_avg, 'b-', label="Ion. In house PIC-MCC")
axs[0, 0].plot(x_bench, y_ni, 'b--', label="Ion LPP")
axs[0, 0].set_xlabel("Time [μs]",fontsize=fontsize)
axs[0, 0].set_ylabel("Norm. Density [-]",fontsize=fontsize)
ax.set_xlim(0, 5)
ax.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))  # or use 'plain' if you prefer
ax.yaxis.get_offset_text().set_fontsize(fontsize)
axs[0, 0].legend(fontsize=18)
axs[0, 0].grid(True)

# (2) Total Energy
axs[0, 1].plot(time_us, total_energy, 'r-', label="Total In house PIC-MCC")
axs[0, 1].plot(bench_time, bench_total_energy, 'b--', label="Total LPP")
axs[0, 1].set_xlabel("Time [μs]", fontsize=fontsize)
axs[0, 1].set_ylabel("Energy [J]", fontsize=fontsize)

ax.set_xlim(0, 5)
ax.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))  # or use 'plain' if you prefer
ax.yaxis.get_offset_text().set_fontsize(fontsize)

axs[0, 1].legend(fontsize=18)
axs[0, 1].grid(True)

# (3) Electron Energy
axs[1, 0].plot(time_us, elec_kin, label="In house PIC-MCC Kinetic")
axs[1, 0].plot(time_us, elec_pot, label="In house PIC-MCC Potential")
axs[1, 0].plot(bench_time, bench_elec_kin, '--', label="LPP Kinetic")
axs[1, 0].plot(bench_time, bench_elec_pot, '--', label="LPP Potential")
axs[1, 0].set_xlabel("Time [μs]", fontsize=fontsize)
axs[1, 0].set_ylabel("Electron Energy [J]", fontsize=fontsize)
ax.set_xlim(0, 5)
ax.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))  # or use 'plain' if you prefer
ax.yaxis.get_offset_text().set_fontsize(fontsize)
axs[1, 0].legend(fontsize=18)
axs[1, 0].grid(True)

# (4) Ion Energy
axs[1, 1].plot(time_us, ion_kin, label="In house PIC-MCC Kinetic")
axs[1, 1].plot(time_us, ion_pot, label="In house PIC-MCC Potential")
axs[1, 1].plot(bench_time, bench_ion_kin, '--', label="LPP Kinetic")
axs[1, 1].plot(bench_time, bench_ion_pot, '--', label="LPP Potential")
axs[1, 1].set_xlabel("Time [μs]", fontsize=fontsize)
axs[1, 1].set_ylabel("Ion Energy [J]", fontsize=fontsize)
ax.set_xlim(0, 5)
ax.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))  # or use 'plain' if you prefer
ax.yaxis.get_offset_text().set_fontsize(20)
axs[1, 1].legend(fontsize=20)
axs[1, 1].grid(True)  


# Global formatting
for i, ax in enumerate(axs.flat):
    ax.set_xlim(0, 5)
    ax.tick_params(axis='both', labelsize=20)
    ax.legend(fontsize=18)
    ax.grid(True)
    for spine in ax.spines.values():
        spine.set_linewidth(2)
    ax.set_title(titles[i], fontsize=20, pad=10)

plt.tight_layout()
plt.savefig("combined_case2_results.png", dpi=300, bbox_inches='tight')
```
